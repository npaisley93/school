You have been given a data set of abalone diameters and weight and a python code that will read in this data and plot it. The plot will also show a model suggested by Manson ( W = 5.1D 5 ).
Compute how well the model fits the observations by computing the SSR. and place this value in a text box on the chart. Run your code a few times changing the values of 5.1 and 5 to other values. Show how the SSR changes.
Use a Python function to compute the SSR as a function of the observed data, and the model parameters.
As before submit your code and result in a pdf.

BASE A=5.1 and BASE B=5.0

```python
#Libraries imported
import matplotlib.pyplot as plt 
from matplotlib import style
import numpy as np
#style.use("ggplot")

def SSR(x,y,a,b): #Creating a definition for SSR where SSR (Squares Residual) is a function of the observed data and model parameters
                  #Residual is the difference between the Observed Y data and what the model predictsfor the Y value at the value of X
    #weight_model = a * x + b ## This would be a linear model
    weight_model = a * x ** b #For the Power Function
    ssr = np.sum((weight_model-y)**2) #sum of [[Y observed - Y prediction]] ^ 2
    return ssr

def model(x,y,a,b):
    #weight_model = a * x + b ## This would be a linear model
    weight_model = a * x ** b #For a Power Function
    return weight_model

filename = "abalone.data.txt"
# x and y coordinates, whitespace-separated
X = np.loadtxt( filename, delimiter=',', usecols=(2,4) ) #column 2 and 4 are the diameter and weights of abalone
diameter = X[:,0] #setting column 1 in X (index 0) to variable "diameter"
weight_observed = X[:,1] #setting column 2 in X (index 1) to variable "weight_observed"

a,b=5.1,5.0 
            #higher the base(a) the closer the obeserved data...makes SSR decrease closer to 0
            #lower the exponent(b) the closer to the observed data...make SSR decrease closer to 0

#Labels for the loss function function SSR and weight model values as variables to be used in label 1 and 2:
loss = SSR(diameter,weight_observed,a,b) #making SSR new variable loss
weight_model = model(diameter,weight_observed,a,b) #making model new variable weight_model

#Model and SSR values/text written above the graph
label1=r'Model: W=' + str(a) + r'D$^{' + str(b) + r'}$' #str=string, str(a) and str(b) are a and b variable as strings
label2=r'SSR = ' + str(loss) #str(loss) is SSR fucntion as a string
print(label1)
print(label2)

#Construction of graph
plt.scatter(diameter,weight_observed, marker = 'o', s=150, alpha = 0.5,label='Observed') #plotting observed values
plt.scatter(diameter,weight_model,marker='.',color='red',label=label1) #plotting model values

plt.text(0.05,2.2,label1) #Text box IN graph below legend listing model and SSR values
plt.text(0.05,2.0,label2) #^^^ numbers accurately postion label in graph out of way of datapoints
plt.xlabel('D') 
plt.ylabel('W')
plt.legend() #offical legend in upper part of graph including the shape/color of points

#plt.savefig("abalone.png")
plt.show()
```
