```python
import numpy as np
import matplotlib.pyplot as plt

# Use a Python function to evaluate a polynomial of degree 5, i.e. a quintic
def poly5(a,x):
    y = a[0] + a[1] * x + a[2] * x**2 + a[3] * x**3 + a[4] * x**4 + a[5] * x**5
    return y

# Use a Python function to numerical estimate the steepness of a 
# polynomial of degree 5, i.e. a quintic.
def steepness5(a,x,dx):
    yup = poly5(a,x+dx)
    ydown = poly5(a,x)
    steepness = (yup-ydown)/dx
    return steepness
    
# Set up a domain from -1.1 to 3.1 with 50 points
x=np.linspace(-1.1,3.1,50)

# Set coefficients of polynomial
a = np.array([2.0,-6.0,5.0,5.0,-5.0,1.0])

# Evaluate polynomial function as y for all X
y = poly5(a,x)

plt.rcParams['figure.figsize'] = [10, 5]
#plt.axis('off')
plt.scatter(x,y)  # Plot the polynomial
plt.xlabel('x')
plt.ylabel('y')
#plt.axes().set_aspect('equal')

xp = 1.5
dx = 0.01
steepness_at_xp = steepness5(a,xp,dx)
print(steepness_at_xp) 
# Plot a short line showing the steepness
#plt.plot([xp, xp+dx], [poly5(a,xp), poly5(a,xp+dx)], 'r-',linewidth=5)
extend_dx = 15
plt.plot([xp, xp+extend_dx*dx], [poly5(a,xp), poly5(a,xp)+steepness_at_xp*extend_dx*dx], 'r-',linewidth=5,alpha=0.5)
plt.plot([xp, xp-extend_dx*dx], [poly5(a,xp), poly5(a,xp)-steepness_at_xp*extend_dx*dx], 'r-',linewidth=5,alpha=0.5)
plt.show()  
```
