for a domain  x = 0 to 30 taking b = 0.05 and with 100 points.

```python
# import libraries that are required
import numpy as np
import sys
import matplotlib.pyplot as plt

x = np.linspace(0,30,100) #from 0 to 30 with 100 points
b = .05
y = 2 + np.exp(-b*x) * np.sin(x) #equation
plt.plot(x,y, color='red', linewidth=5)
plt.xlim([0,30]) #x domain
plt.ylim([0,4]) #y domain
plt.xlabel('x',fontsize=18) #x label
h=plt.ylabel('y',fontsize=18) #y label
plt.title('$y = 2 + e^{(-.05x)} * sin{(x)}$',fontsize=18) #title of graph
h.set_rotation(0) #rotates y 
plt.grid()

plt.show()
```

The intercept is at 2, so that is where the graph starts. Since it is a since graph it is going to be a wave.
The reason the amplitude is like this is because of the exponent. The smaller the exponent is the more waves are created.
Since this exponent is .05, it is going to have more waves than .1.
